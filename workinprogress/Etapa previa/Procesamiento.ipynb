```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataK = pd.read_csv('./Konex/data.csv')
columnas = ['Formacion','Justificacion','Seguridad','Truco']
N = len (columnas)
```

```python
f, axes = plt.subplots(1, N, figsize=(7*N, 7), sharey=True)
count = 0
for columna in columnas:
    if dataK[columna].dtype == 'float64':
        g = sns.distplot (dataK[columna], ax=axes[count])
    else:
        g = sns.countplot (dataK[columna], ax=axes[count])
        g.set_xticklabels(g.get_xticklabels(), rotation=60)
        count +=1
```

```python
data = dataK[columnas]
data.head()
sns.pairplot(data)
```
