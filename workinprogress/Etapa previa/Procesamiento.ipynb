```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./RAFA 2016/data.csv')
columnas = data.columns
#columnas = ['Formacion','Justificacion','Seguridad','Truco']
N = len (data.columns)
```

```python
data.columns
```

```python
f, axes = plt.subplots(1, N, figsize=(7*N, 7), sharey=True)
count = 0
for columna in columnas:
    if data[columna].dtype == 'float64':
        g = sns.distplot (data[columna], ax=axes[count])
    else:
        g = sns.countplot (data[columna], ax=axes[count])
        g.set_xticklabels(g.get_xticklabels(), rotation=60)
        count +=1
plt.savefig('prueba.png')
```
